```python
# Importing the required libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from keras.layers import Dense        # How many hidden layers we want to create is known as Dense
from keras.models import Sequential   # Neural Networks is conducted insequence
import tensorflow as tf
```

```python
# Read the dataset and before it doesn't have column names and we gave the column names as below.

# Where, class is dependent variable and rest of the columns is independent variable

n="preg","plas","pres","skin","test","mass","pedi","age","class"
df=pd.read_csv("/content/pima-indians-diabetes.data.csv",names=n)
```

```python
df
```

```python
array=df.values
```

```python
array
```

```python
# Splitting the data into Independent Variable is X & Dependent Variable is Y

X=array[:,0:8]
y=array[:,8]
```

```python
X
```

```python
y
```

```python
df.describe()
```

```python
# Building the Keras Model

model=Sequential()
model.add(Dense(12,input_dim=8,activation="relu"))    # In Dense we have 12 hidden layers,input_dim we have 8 neurons layers because we have 8 independent variables and Activation function is ReLu
model.add(Dense(8,activation="relu"))                 # We have 8 hidden layers because we have 8 independent variables and Activation function is ReLu
model.add(Dense(1,activation="sigmoid"))              # Output layer is 1 and activation function is Sigmoid.
```

```python
# Compile The Keras Model
# It is classification based problem so, we used the binary crossentropy.
# What it is regression based problem so, we can use the RMSE, MSE.

model.compile(loss="binary_crossentropy",optimizer="adam",metrics=["accuracy"])
```

```python
# Fitting the model in the data
# Batch size is 10 - So, it Batch Gradient Descent Model
# Batch size is 1 - Then it is Stochastic Gradient Descent Model
# Batch Size is 1 to n - Then it is Mini Batch Gradient Descent Model
# Epochs =150 it means time

history=model.fit(X,y,epochs=150,validation_split=0.33,batch_size=10)
```

```python
# Evauluating the keras model

scores=model.evaluate(X,y)
```

```python
# Score Value

print("%s: %.2f%%" % (model.metrics_names[1],scores[1]*100))
```

```python
# Plotting the accuracy Rate

plt.plot(history.history["accuracy"])
plt.plot(history.history["val_accuracy"])
plt.title("Model Accuracy")
plt.ylabel("Accuracy")
plt.xlabel("Epoch")
plt.legend(["train","test"],loc="upper left")
plt.show()

# Plotting the loss rate

plt.plot(history.history["loss"])
plt.plot(history.history["val_loss"])
plt.title("Model Loss")
plt.ylabel("Loss")
plt.xlabel("Epoch")
plt.legend(["train","test"],loc="upper left")
plt.show()
```

  By observing the above insights we have some scope to pushup the accuracy.

  By experimenting the below hyperparameters we can pushup the accuracy score.
- Change the epoch values.
- Change the Activation Function.
- Change the number of neurons.
- Change the number of hidden layers.
- Change the optimizers
